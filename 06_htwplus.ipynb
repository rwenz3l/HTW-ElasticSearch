```bash
# Ubuntu 16.04.2 LTS VM in VMWare Fusion 8
# Installation von Ubuntu Server durchführen
# PostgreSQL wird schon bei Installation angeboten <-- Auswählen
# OpenSSH Server für den Zugriff via SSH <-- Auswählen
```

```bash
# Installation Abgeschlossen

## User = elastic
## Pass = elastic

## Root-PW = elasticfantastic
# In /etc/ssh/sshd_config
# PermitRootLogin yes
# Sonst kann man sich nicht anmelden

service sshd status
# Should be running
service sshd restart

service postgresql status
# Should be running as well

ip a
# Note the IP-Address
# 172.16.253.128
ssh elastic@172.16.253.128
# ALternativ
ssh root@172.16.253.128

sudo su
# Let's install some stuff
apt update

apt install python-software-properties
add-apt-repository ppa:webupd8team/java
apt update
apt install oracle-java8-installer
# Accept and Install

# Set Firewall rules
ufw allow 9000/tcp
ufw allow 22/tcp
ufw enable

#Download ElasticSearch

cd /tmp && \
wget https://download.elastic.co/elasticsearch/release/org/elasticsearch/distribution/deb/elasticsearch/2.4.5/elasticsearch-2.4.5.deb

dpkg -i elasticsearch-2.4.5.deb

# Check Status (Should be off)
service elasticsearch status

# Start Service
service elasticsearch start

# Check Status (Should be on now)
service elasticsearch status

rm elasticsearch-2.4.5.deb

# Kopiere die Uebung06 via SFTP/SSH/etc auf die VM (home ordner)
apt install unzip
cd /home/elastic
unzip Uebung6.zip

mv Uebung6/htwplus /opt/

# As Root:
su postgres
psql
# In PSQL:
postgres=# CREATE USER elastic PASSWORD 'fantastic';
CREATE ROLE
postgres=# CREATE DATABASE elastic OWNER='elastic';
CREATE DATABASE
quit;

# In Shell, as root:
nano /etc/postgresql/9.5/main/pg_hba.conf
# Change
local   all             all                                     peer
# to
local   all             all                                     md5
# Save and Quit

# Try to login
psql -d elastic -U elastic
# login with pw = fantastic
# In PSQL:
# Import Database:
elastic=> \i /home/elastic/Uebung6/htwplus_test_data.sql
# Quit
elastic=> \q

# Workon HTWPlus
cd /opt/htwplus/
nano conf/application.conf

# Set Values in application.conf
db.default.driver=org.postgresql.Driver
db.default.url="jdbc:postgresql://localhost/elastic"
db.default.username=elastic
db.default.password=fantastic
# [..]
media.path="/opt/htwplus/public/media"
media.tempPath="/opt/htwplus/public/media/tmp"
# [..]
media.fileStore="/opt/htwplus/public/media/filestore"

# Starten!
chmod +x activator
./activator run

# Öffnen:
http://172.16.253.128:9000
# Anmelden mit admin@htwplus.de

# WIP NFS for Working on this:
apt install nfs-kernel-server
echo '/opt/htwplus *(rw,async,insecure,no_subtree_check,all_squash)' >> /etc/exports
service nfs-kernel-server reload

ufw allow 111
ufw allow 2049
ufw reload
# Or just disable since it also requires a random port:
ufw disable
# Connect via NFS:
nfs://172.16.253.128/opt/htwplus
```
